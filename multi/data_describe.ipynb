```python
# Training Set
train_df = pd.read_hdf('./train_multi_inputs.h5', start= 0, stop = 1)
train_df = train_df.reset_index(level=0)
train_df = train_df['cell_id']
train_df = pd.merge(train_df, metadata, how='left',on='cell_id')

for i in range(600):
    temp = pd.read_hdf('./train_multi_inputs.h5', start= 1024 * i + 1, stop=1024 * i + 1024)
    temp = train.reset_index(level=0)
    temp = train['cell_id']

    temp = pd.merge(temp, metadata, how='left',on='cell_id')

    train_df = pd.concat([train_df, temp], ignore_index=True)
```
